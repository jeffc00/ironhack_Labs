```python
#------- Getting the coordinates ----- 
```

```python
def coordinates(x):
    time.sleep(5)
    try:
        x = str(x).replace(' ','+')
        print(x)
        url = f'https://nominatim.openstreetmap.org/search?q={x}&format=geojson'
        d = requests.get(url).json()
        try:
            return (d['features'][0]['geometry']['coordinates'][1], d['features'][0]['geometry']['coordinates'][0])
        except:
            print(f'Localização <{x}> não obtida!')
            return np.nan
    except:
        pass
```

```python
terrorists['Coordinates'] = terrorists['Place of Birth'].apply(lambda x : coordinates(x))
```

```python
terrorists['Coordinates'].fillna(0, inplace=True)
```

```python
n = []
c = []
for i,j in zip(terrorists['Name'],terrorists['Coordinates']):
    if j != 0:
        n.append(i)
        c.append(j)
```

```python
url = 'https://boards.greenhouse.io/nubank?gh_src=b58135231'
```

```python
nu = BeautifulSoup(requests.get(url).content)
```

```python
# Procurando jobs at Nubank
```

```python
jobs_name = [item.a.text for item in nu.body.find_all('div',{'class':'opening'})]
jobs_location = [item.span.text for item in nu.body.find_all('div',{'class':'opening'})]
```

```python
n = len(jobs_name)
d = {}
for i in range(n):
    e = {}
    e['Posicao'] = jobs_name[i]
    e['Localidade'] = jobs_location[i]
    d[i] = e
```

```python
jobsAtnu = pd.DataFrame(d).T
```

```python
jobsAtnu['Posicao'].apply(lambda x : re.findall(r'.*[dD]ata.*',x)).sum()
```
