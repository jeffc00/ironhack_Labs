```python
url = 'https://boards.greenhouse.io/nubank?gh_src=b58135231'
```

```python
nu = BeautifulSoup(requests.get(url).content)
```

```python
# Procurando jobs at Nubank
```

```python
jobs_name = [item.a.text for item in nu.body.find_all('div',{'class':'opening'})]
jobs_location = [item.span.text for item in nu.body.find_all('div',{'class':'opening'})]
```

```python
n = len(jobs_name)
d = {}
for i in range(n):
    e = {}
    e['Posicao'] = jobs_name[i]
    e['Localidade'] = jobs_location[i]
    d[i] = e
```

```python
jobsAtnu = pd.DataFrame(d).T
```

```python
x = jobsAtnu.groupby('Localidade').count()
```

```python
x.reset_index(inplace=True)
```

```python
sns.barplot(x= x['Localidade'], y=x['Posicao']);
```
