```python
## Game number one: Guess The Number

import random

number = int(random.random()*100)

name = input('Digite seu nome: \n')
print(f'\n\nBem vindo {name}, ao Guess the Number:\n\n')

print('''OBJETIVO DO JOGO:\nVocê deverá adivinhar em 5 tentivas um número que foi gerado de forma aleatória de 1 a 100.\n
REGRAS DO JOGO:
A) Se você estiver a mais que 20 unidades do número aparecerá na tela: "GELADO" 
B) menos que 20, porém mais que 10: "FRIO" 
B) menos de 10, porém mais que 5: "MORNO"
C) menos que 5: "QUENTE"''')

while True:
    try:
        tries = int(input('Quantas tentivas? '))
        break
    except:
        print('Por favor digite um número inteiro!')

print(number)
game = 1
while True:
    
    while True:
        try:
            n_user = int(input('Digite o número da sorte entre 1 e 100: '))
            if n_user < 0 or n_user >100:
                print("Por favor digite um número entre 1 e 100! ")
                continue
            else:
                break
        except:
            print('Por favor digite um número inteiro!')
    n = abs(n_user - number)
    if n == 0:
        print('Parabés, Você acertou o número!')
        break
    else:
        if n >= 20:
            print(f'\nTentativa : {game} --> GELADO')
        elif n < 20 and n >= 10:
            print(f'\nTentativa : {game} --> FRIO')
        elif n >= 5 and n < 10:
            print(f'\nTentativa : {game} --> MORNO')
        elif n < 5:
            print(f'\nTentativa : {game} --> QUENTE')
    game += 1
    if (game > tries):
        q = input('Ops!!\nParece que você perdeu :(\nDeseja tentar de novo? [s][n] ')
        if q.lower() == 's':
            game = 1
            continue
        else:
            print("\nTchau!")
            break
    
```

```python
#Caesar Cipher
#32 = 100000

def crypto(string, d = 3):
    cryp = ''
    for i in string:
        #print(ord(i))
        c = ord(i) + d
        if ord(i) == 32:
            cryp += i
        elif i.isupper():
            #range 65 - 90
            c = (c-65)%26 + 65
            cryp += chr(c)
        elif i.islower():
            #range 97 - 122
            c = (c-97)%26 + 97
            cryp += chr(c)
        else:
            cryp += chr(c)
    
    return cryp
                
def decrypto(string, d = 3):
    cryp = ''
    for i in string:
        c = ord(i) - d
        if ord(i) == 32:
            cryp += i
        elif i.isupper():
            c = (c-65)%26 + 65
            cryp += chr(c)
        elif i.islower():
            c = (c-97)%26 + 97
            cryp += chr(c)
        else:
            cryp += i
            
            
    return cryp

choice = int(input('Digite 1 para criptografia ou 2 para decriptografia:'))
string = str(input('Digite o texto para criptografia: '))
d = int(input('Digite o shift: '))
if choice == 1:
    print(f'\n{crypto(string,d)}')
else:
    print(f'\n{decrypto(string,d)}')
```

```python
def forca(n):
    if n == 1:
        print('o')
    elif n==2:
        print('o\n|')
    elif n==3:
        print(' o\n/|\'')
    elif n==4:
        print(' o\n/|\\\n')
    elif n==5:
        print(' o\n/|\\\n |')
    elif n==6:
        print(' o\n/|\\\n |\n/')
    elif n==7:
        print(' o\n/|\\\n |\n/ \\')
    elif n==8:
        print(' o\n/|\\\n |\n/ \\\n')

list = ['pandas', 'python', 'numpy', 'sql', 'postgresql', 'csv', 'ironhack', 'mysql']

n = random.randint(0,7)

word_guess = list[n]
length = len(word_guess)

empty = []

for i in range(length):
    empty.append('_')
    
def print_hang():
    print(' '.join(empty))

print(word_guess)
print_hang()
tentativas = 12
game = 1
fork = 0
while True:
    temp = False
    
    
    if '_' not in empty:
        print('\nParabéns! Vc ganhou')
        break
    
    s = input('\nDigite uma letra ')
    
    for i in range(length):
        if s == word_guess[i]:
            empty[i] = s
            temp = True
        else:
            continue
    if temp == False:
        fork+=1
        if fork > 8:
            print('Você perdeu!')
            break
    print('\n\n')
    forca(fork) 
    print('\n')
    print_hang()
    game+=1
    if game>tentativas:
        print('Você perdeu!!')
        break
```
