```python
def rob_string(x):
    if re.findall(r'.*B\.C*.', x):
        return int(''.join(re.findall(f'\d+', x)))*-1
    elif re.findall(r'.*[bB]efore.*', x):
        return ''.join(re.findall(f'\d+', x))
    else:
        return 0 
```

```python
rob_string(texto)
```
